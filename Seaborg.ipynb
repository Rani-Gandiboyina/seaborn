```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
df=sns.load_dataset("tips")
```

```python
sns.stripplot(y='tip',data=df,jitter=True)

plt.ylabel('tips($)')
```

```python
sns.stripplot(x="day",y="tip",data=df)
```

```python
sns.stripplot(x="day",y="tip",data=df,size=4,jitter=False)
```

```python
sns.swarmplot(x='day',y='tip',data=df)
```

```python
sns.swarmplot(x='day',y='tip',data=df,hue='sex')
```

```python
sns.swarmplot(x='tip',y='day',data=df,hue='sex',orient='h')
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.boxplot(x='day',y='tip',data=df)
```

```python
sns.violinplot(x='day',y='tip',data=df)
```

```python
plt.subplot(1,2,1)

sns.boxplot(x='day',y='tip',data=df)
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=df)
```
